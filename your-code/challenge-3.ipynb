You are serving as a Business Intelligence Analyst at the headquarter of an international fashion goods chain store. Your boss today asked you to do two things for her:

**First, identify two groups of customers from the data set.** The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). The second group is **Preferred Customers** whose **aggregated expenses** are **between the 75th and 95th percentile**.

**Second, identify which country has the most of your VIP customers, and which country has the most of your VIP+Preferred Customers combined.**

## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries

import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
orders_df= pd.read_csv('orders.csv')
orders_df.head()
```

---

"Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1:

# Grouping by CustomerID and sum the amount_spent for each customer:

customer_expenses = orders_df.groupby('CustomerID')['amount_spent'].sum()

customer_expenses
```

```python
# Sub Problem 2:

# Calculating the 75th and 95th percentiles:

percentile_75 = customer_expenses.quantile(0.75)
percentile_95 = customer_expenses.quantile(0.95)

# Selecting Preferred Customers (between 75th and 95th percentile):

preferred_customers = customer_expenses[(customer_expenses >= percentile_75) & (customer_expenses < percentile_95)]

# Selecting VIP Customers (above 95th percentile):

vip_customers = customer_expenses[customer_expenses >= percentile_95]

# Calculating the minimum and maximum spending for Preferred and VIP customers:

min_spending_preferred = round(preferred_customers.min(),2)
max_spending_preferred = round(preferred_customers.max(), 2)

min_spending_vip = round(vip_customers.min(), 2)
max_spending_vip = round(vip_customers.max(), 2)


print("Range for Preferred Customers: Min =", min_spending_preferred, "Max =", max_spending_preferred)
print("Range for VIP Customers: Min =", min_spending_vip, "Max =", max_spending_vip)
```

```python
# Sub Problem 3:

# Creating a new column 'CustomerCategory' and adding default values first as Regular:

orders_df['CustomerCategory'] = 'Regular'

# Defining Preferred Customers:

orders_df.loc[orders_df['CustomerID'].isin(preferred_customers.index), 'CustomerCategory'] = 'Preferred'

# Defining VIP Customers:

orders_df.loc[orders_df['CustomerID'].isin(vip_customers.index), 'CustomerCategory'] = 'VIP'

orders_df.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# Since we already defined variables on the precious solution, we proceed directly as follows:

# Grouping VIP Customers by country and counting them:

vip_customers_by_country = orders_df[orders_df['CustomerID'].isin(vip_customers.index)]

vip_customers_count_by_country = vip_customers_by_country['Country'].value_counts()

# Finding  the country with the most VIP Customers:

country_with_most_vip = vip_customers_count_by_country.idxmax()


print("Country with the most VIP Customers:", country_with_most_vip)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Calculating the count of Preferred Customers by grouping and counting orders (vip already done Q2):

preferred_customers_by_country = orders_df[orders_df['CustomerID'].isin(preferred_customers.index)]
preferred_customers_count_by_country = preferred_customers_by_country['Country'].value_counts()

# Creating a DF with counts for both Customers:

vip_preferred_counts_by_country = pd.DataFrame({'VIP': vip_customers_count_by_country,'Preferred': preferred_customers_count_by_country})

# Calculating the count of VIP+Preferred Customers:

vip_preferred_counts_by_country['VIP+Preferred'] = vip_preferred_counts_by_country['VIP'] + vip_preferred_counts_by_country['Preferred']

# Find the country with the most VIP+Preferred Customers
country_with_most_vip_preferred = vip_preferred_counts_by_country['VIP+Preferred'].idxmax()

print("Country with the most VIP+Preferred Customers:", country_with_most_vip_preferred)
```
